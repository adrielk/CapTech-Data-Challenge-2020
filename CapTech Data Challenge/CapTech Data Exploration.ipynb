```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import pylab as pl
sns.set(style = 'darkgrid')
#matplotlib inline
```

```python
gender_edu_df = pd.read_csv('gender_education_2010-2019.csv')
age_gender_df = pd.read_csv('age_gender_2010-2019.csv')
race_eth_df = pd.read_csv('race_ethnicity_2010-2019.csv')
crime_df = pd.read_csv('Crime Stats by State_2005-2014.csv')
```

```python
crime_df.head()
```

```python
gender_edu_df.head()
```

```python
gE_state= gender_edu_df[['State','beg employment','end employment','avg mthly earnings','qrtly payroll']]
gE_industry =  gender_edu_df[['industry','beg employment','end employment','avg mthly earnings','qrtly payroll']]
```

```python
gE_state_agg = gE_state.groupby(["State"]).median().reset_index()
gE_industry_agg = gE_industry.groupby(["industry"]).median().reset_index()
```

```python
gE_state_agg.head()
```

```python
gE_industry_agg.head()
```

## Average Monthly Earnings and Quarterly Payrolls by State
Notes: 
- I'm not sure what Quarterly payroll is...
- There are 4 quarters, is it safe to just average across all quarters?

```python
state_vs_earnings = gE_state_agg[["State","avg mthly earnings"]]

state_vs_earnings = state_vs_earnings.sort_values(by=['avg mthly earnings'])

state = state_vs_earnings["State"]
earnings = state_vs_earnings["avg mthly earnings"]

plt.figure(figsize = (10,10))
plt.title("Median of Average of Monthly Earnings of Each State")
plt.xlabel("State")
plt.ylabel("Monthly Earnings (USD)")
plt.xticks(rotation='vertical')
plt.barh(state,earnings)
```

```python
state_vs_payroll = gE_state_agg[["State","qrtly payroll"]]

state_vs_payroll = state_vs_payroll.sort_values(by=['qrtly payroll'])

state = state_vs_payroll["State"]
payroll = state_vs_payroll["qrtly payroll"]

plt.figure(figsize = (10,10))
plt.title("Median Quarterly by State")
plt.xlabel("State")
plt.ylabel("Average Quarterly Payroll (USD)")
plt.xticks(rotation='vertical')
plt.barh(state,payroll)
```

## Median Monthly Earnings and Quarterly Payrolls by Industry


```python
industry_vs_earnings = gE_industry_agg[["industry","avg mthly earnings"]]

industry_vs_earnings = industry_vs_earnings.sort_values(by=['avg mthly earnings'])

industry = industry_vs_earnings["industry"]
earnings = industry_vs_earnings["avg mthly earnings"]

plt.figure(figsize = (10,10))
plt.title("Median of Average of Monthly Earnings of Each Industry")
plt.xlabel("Industry")
plt.ylabel("Monthly Earnings (USD)")
plt.xticks(rotation='vertical')
plt.barh(industry,earnings)
```

```python
industry_vs_payroll = gE_industry_agg[["industry","qrtly payroll"]]

industry_vs_payroll = industry_vs_payroll.sort_values(by=['qrtly payroll'])

industry = industry_vs_payroll["industry"]
payroll = industry_vs_payroll["qrtly payroll"]

plt.figure(figsize = (10,10))
plt.title("Median of Quarterly Payroll of Each Industry")
plt.xlabel("Industry")
plt.ylabel("Quarterly Payroll (USD)")
plt.xticks(rotation='vertical')
plt.barh(industry,payroll)
```

## Payroll over Time by National Median, then by each State each Industry
- Have the average trend, then individual trends for top locations and top industries
- Some how over lay the trend lines....?

```python
#Average trend...
time_payroll = gender_edu_df[['year','qrtly payroll']].groupby("year").median().reset_index()
time_mthly_earnings = gender_edu_df[['year','avg mthly earnings']].groupby("year").median().reset_index()
```

```python
time_payroll.head()
```

```python
time_mthly_earnings.head()
```

```python
year = time_payroll["year"]
payroll = time_payroll["qrtly payroll"]

plt.figure(figsize = (10,10))
plt.title("Overall National Median Quartery Payroll over Time ")
plt.xlabel("Time (Year)")
plt.ylabel("Quarterly Payroll (USD)")
plt.xticks(rotation='vertical')
plt.plot(year,payroll)
```

```python
year = time_mthly_earnings["year"]
earnings = time_mthly_earnings["avg mthly earnings"]

plt.figure(figsize = (10,10))
plt.title("Overall National Average Monthly Earnings over Time ")
plt.xlabel("Time (Year)")
plt.ylabel("Avg Monthly Earnings (USD)")
plt.xticks(rotation='vertical')
plt.plot(year,earnings)
```
