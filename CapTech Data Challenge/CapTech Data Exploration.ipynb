```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import pylab as pl
sns.set(style = 'darkgrid')
#matplotlib inline
```

```python
gender_edu_df = pd.read_csv('gender_education_2010-2019.csv')
age_gender_df = pd.read_csv('age_gender_2010-2019.csv')
race_eth_df = pd.read_csv('race_ethnicity_2010-2019.csv')
crime_df = pd.read_csv('Crime Stats by State_2005-2014.csv')
```

```python
gE_state= gender_edu_df[['State','beg employment','end employment','avg mthly earnings','qrtly payroll']]
gE_industry =  gender_edu_df[['industry','beg employment','end employment','avg mthly earnings','qrtly payroll']]
```

```python
gE_state_agg = gE_state.groupby(["State"]).median().reset_index()
gE_industry_agg = gE_industry.groupby(["industry"]).median().reset_index()
```

## Average Monthly Earnings and Quarterly Payrolls by State

```python
state_vs_earnings = gE_state_agg[["State","avg mthly earnings"]]

state_vs_earnings = state_vs_earnings.sort_values(by=['avg mthly earnings'])

state = state_vs_earnings["State"]
earnings = state_vs_earnings["avg mthly earnings"]

plt.figure(figsize = (10,10))
plt.title("Median of Average of Monthly Earnings of Each State")
plt.xlabel("State")
plt.ylabel("Monthly Earnings (USD)")
plt.xticks(rotation='vertical')
plt.barh(state,earnings)
```

```python
state_vs_payroll = gE_state_agg[["State","qrtly payroll"]]

state_vs_payroll = state_vs_payroll.sort_values(by=['qrtly payroll'])

state = state_vs_payroll["State"]
payroll = state_vs_payroll["qrtly payroll"]

plt.figure(figsize = (10,10))
plt.title("Median Quarterly by State")
plt.xlabel("State")
plt.ylabel("Average Quarterly Payroll (USD)")
plt.xticks(rotation='vertical')
plt.barh(state,payroll)
```

```python
#Pivot table source (very nice): https://stackoverflow.com/questions/30679467/pivot-tables-or-group-by-for-pandas/30679543

#Groupby state/industry, and year and average qrtly payroll and avg mthly earnings
state_time_payroll = gender_edu_df[['State','year','qrtly payroll']]#.groupby(['State','year']).median().reset_index()
state_time_payroll_pivot = pd.pivot_table(state_time_payroll, values=['qrtly payroll'], index=['State'],columns = ['year'],aggfunc={'qrtly payroll': np.median}).reset_index()
state_time_payroll_pivot.head()
```

```python
state_time_earnings = gender_edu_df[['State','year','avg mthly earnings']]#.groupby(['State','year']).median().reset_index()
state_time_earnings_pivot = pd.pivot_table(state_time_earnings, values=['avg mthly earnings'], index=['State'],columns = ['year'],aggfunc={'avg mthly earnings': np.median}).reset_index()
state_time_earnings_pivot.head()
```

```python
industry_time_payroll = gender_edu_df[['industry','year','qrtly payroll']]#.groupby(['industry','year']).median().reset_index()
industry_time_payroll_pivot = pd.pivot_table(industry_time_payroll, values=['qrtly payroll'], index=['industry'],columns = ['year'],aggfunc={'qrtly payroll': np.median}).reset_index()
industry_time_payroll_pivot.head()
```

```python
industry_time_earnings = gender_edu_df[['industry','year','avg mthly earnings']]#.groupby(['industry','year']).median().reset_index()
industry_time_earnings_pivot = pd.pivot_table(industry_time_earnings, values=['avg mthly earnings'], index=['industry'],columns = ['year'],aggfunc={'avg mthly earnings': np.median}).reset_index()
industry_time_earnings_pivot.head()
```

### Yearly Quarterly Payroll for 6 Top States:
- NOTE: Top States chosen based on greatest median quarterly payroll

```python
plt.figure(figsize = (10,10))

year_cols = list(state_time_payroll_pivot.columns.get_level_values(1))[1:]

#:( I couldn't figure out how to sort by column for a pivot table....
#These top states were selected based on largest median quarterly payroll
top_states = ['California','Texas','New York','Florida','Illinois','Pennsylvania']


for i in range(len(list(state_time_payroll_pivot["State"]))):
    #print(state_time_payroll_pivot.iloc[i])
    current_state = state_time_payroll_pivot["State"][i]
    if(current_state in top_states):
        state_row_payroll = list(state_time_payroll_pivot.iloc[i]['qrtly payroll'])


        plt_title = "Yearly Quarterly Payroll For Top States"
        plt.title(plt_title)
        plt.xlabel("Time (Year)")
        plt.ylabel("Quarterly Payroll (USD)")
        plt.xticks(rotation='vertical')
        plt.plot(year_cols,state_row_payroll,label=str(current_state))
        
plt.plot(year_cols,payroll, label = "Overall Median")
plt.legend(loc=3,bbox_to_anchor=(1,0))
```
